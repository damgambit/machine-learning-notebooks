# Load data

```python
dtype_dict = {'bathrooms':float, 
              'waterfront':int, 
              'sqft_above':int, 
              'sqft_living15':float, 
              'grade':int, 
              'yr_renovated':int, 
              'price':float, 
              'bedrooms':float, 
              'zipcode':str, 
              'long':float, 
              'sqft_lot15':float, 
              'sqft_living':float, 
              'floors':str, 
              'condition':int, 
              'lat':float, 
              'date':str, 
              'sqft_basement':int, 
              'yr_built':int, 
              'id':str, 
              'sqft_lot':int, 
              'view':int
             }

train_data = pd.read_csv('kc_house_train_data.csv', dtype=dtype_dict)
test_data = pd.read_csv('kc_house_test_data.csv', dtype=dtype_dict)
```

# Simple Linear Regression Model
## Using Closed-form solution (Gradient = 0)

```python
def simple_linear_regression(input_feature, output):
    # Useful params
    N = output.shape[0]
    Y = np.sum(output, axis=0)
    X = np.sum(input_feature, axis=0)
    YX = np.dot(output, input_feature)
    X2 = np.dot(input_feature, input_feature)
    
    #Compute slope and intercept
    slope = (YX - (Y*X)/N) / (X2 -  (X*X)/N)
    intercept = Y/N - slope*(X/N)
    return intercept, slope
```

# Function to calculate the Residual Sum of Squares (RSS)

```python
def get_residual_sum_of_squares(input_feature, output, intercept,slope):
    term1 = output - (intercept + slope*input_feature)
    RSS = np.dot(term1, term1)
    return RSS
```

# Define Input and Output columns

```python
input_feature = train_data['sqft_living']
output = train_data['price']
```

# Compute the intercept and the slope using the simple linear regression model

```python
intercept, slope = simple_linear_regression(input_feature, output)
print "Intercept:", intercept
print "Slope:", slope
```

# Predictions

```python
def get_regression_predictions(input_feature, intercept, slope):
    predicted_output = intercept + input_feature * slope
    return predicted_output
```

```python
predictions = get_regression_predictions(input_feature, intercept, slope)
```

```python
plt.plot(train_data['sqft_living'],train_data['price'],'.',
        train_data['sqft_living'],predictions,'-')
```

# Predicted price for a house with 2650 sqft

```python
print "Predicted price for a 2650 sqft house:", get_regression_predictions(2650, intercept, slope)
```

# Compute the RSS

```python
print "RSS:", get_residual_sum_of_squares(input_feature, output, intercept,slope)
```

# Inverse Regression Predictions

```python
def inverse_regression_predictions(output, intercept, slope):
    estimated_input = (output - intercept) / slope
    return estimated_input
```

### Inverse regression prediction for a $800,000 hose

```python
print "Estimated sqft for a $800,000 house:", inverse_regression_predictions(800000, intercept, slope)
```

# Train the model on # of bedrooms instead of sqft_living

```python
intercept_bedrooms, slope_bedrooms = simple_linear_regression(input_feature_bd, output)
print "Intercept:", intercept_bedrooms
print "Slope:", slope_bedrooms
```

```python
rss_sqft = get_residual_sum_of_squares(test_data['sqft_living'], test_data['price'], intercept,slope)
rss_bedrooms = get_residual_sum_of_squares(test_data['bedrooms'], test_data['price'], intercept_bedrooms,slope_bedrooms)
```
